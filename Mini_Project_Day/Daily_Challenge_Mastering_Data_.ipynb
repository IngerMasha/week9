```python
import seaborn as sns

iris = sns.load_dataset('iris')
print(iris.head())
print(iris.info())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(iris['petal_length'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Petal Lengths')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

sns.pairplot(iris, hue='species', palette='husl')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

```python
from plotnine import ggplot, aes, geom_point, labs, theme, element_text

plot_scatter = (
    ggplot(iris, aes(x='sepal_length', y='sepal_width', color='species')) +
    geom_point(size=3, alpha=0.7) +
    labs(title='Sepal Length vs. Sepal Width', x='Sepal Length', y='Sepal Width') +
    theme(axis_title_x=element_text(size=12), axis_title_y=element_text(size=12), plot_title=element_text(size=14, weight='bold'))
)
print(plot_scatter)
```
