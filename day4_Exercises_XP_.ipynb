Exercise 1 : Basic Scatter Plot With Plotnine



```python
import pandas as pd
from plotnine import *
import numpy as np

np.random.seed(0)
data = {
    'x': np.random.rand(20),
    'y': np.random.rand(20)
}
df = pd.DataFrame(data)

scatter_plot = (
    ggplot(df, aes(x='x', y='y')) +
    geom_point(color='blue') +
    labs(x='X-Value', y='Y-Value', title='Basic Scatter Plot') +
    theme_minimal()
)

print(scatter_plot)
```

Exercise 2 : Bar Plot Of Categorical Data With Plotnine

```python
categories = ['A', 'B', 'C']
np.random.seed(1)
data = {
    'category': np.random.choice(categories, 30)
}
df = pd.DataFrame(data)

bar_plot = (
    ggplot(df, aes(x='category')) +
    geom_bar(fill='pink') +
    labs(x='Category', y='Frequency', title='Category Frequency') +
    theme_minimal() +
    theme(axis_text_x=element_text(angle=45, hjust=1))
    )

print(bar_plot)
```

Exercise 3 : Line Plot With Multiple Lines With Plotnine

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
df = pd.DataFrame({'x': x, 'y1': y1, 'y2': y2})

line_plot = (
    ggplot(df, aes(x='x')) +
    geom_line(aes(y='y1'), color='lightblue', size=1.5, linetype='solid') +
    geom_line(aes(y='y2'), color='pink', size=1.5, linetype='dashed') +
    labs(x='X-Value', y='Y-Value', title='Comparison of Two Functions') +
    scale_color_manual(values=['lightblue', 'pink']) +  theme_minimal()
)

print(line_plot)
```

Exercise 4 : Basic Line Chart With Plotly

```python
import plotly.graph_objects as go

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [200, 210, 250, 300, 320, 340, 360, 380, 390, 400, 420, 440]
fig = go.Figure()
fig.add_trace(go.Scatter(x=months, y=sales, mode='lines+markers', name='Sales'))
fig.update_layout(title='Monthly Sales Data', xaxis_title='Month', yaxis_title='Sales')
fig.show()
```

Exercise 5 : Bar Chart With Customization With Plotly

```python
genres = ['Fiction', 'Non-Fiction', 'Sci-Fi', 'Mystery', 'Biography']
books_sold = [120, 150, 90, 80, 130]
fig = go.Figure()
fig.add_trace(go.Bar(x=genres, y=books_sold, marker_color='pink'))
fig.update_layout(title='Books Sold by Genre', xaxis_title='Genres', yaxis_title='Books Sold')
fig.show()
```

Exercise 6 : Scatter Plot With Size And Color Dimensions With Plotly

```python
import plotly.express as px

data = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000],
    'Years of Experience': [2, 5, 7, 10, 12]
})

fig = px.scatter(data, x='Age', y='Salary', size='Years of Experience', color='Years of Experience',
                 title='Relationship between Age, Salary, and Experience',
                 labels={'Age': 'Age (years)', 'Salary': 'Salary ($)', 'Years of Experience': 'Experience (years)'})

fig.show()
```
